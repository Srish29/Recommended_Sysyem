```python
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns 

%matplotlib inline
```

```python
dat_movies = 'movies.dat'
user = pd.read_csv(dat_movies, sep='\t', encoding ='latin-1')
del user['imdbID']
del user['spanishTitle']
del user['imdbPictureURL']
del user['rtID']
user.head()

ratings = user[['id', 'title']].copy()
ratings.rename(columns = {"id": "movieID"}, inplace = True) 
ratings.head()
```

```python
dat_rate = 'user_ratedmovies-timestamps.dat'
rate = pd.read_csv(dat_rate, sep='\t', encoding ='latin-1')
rate.head()
```

```python
ratings = pd.merge(ratings,rate,on='movieID')
ratings.head()
```

```python
ratings.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
ratings.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rat = pd.DataFrame(ratings.groupby('title')['rating'].mean())
rat["count"] = ratings.groupby('title')['rating'].count()
rat.head()
```

```python
moviemat = ratings.pivot_table(index='userID',columns='title',values='rating')
moviemat.head()
```

```python
a = moviemat["'Crocodile' Dundee II"]
a.head()
```

```python
b= moviemat.corrwith(a)
b.head()
```

```python
corr = pd.DataFrame(b,columns=['Correlation'])
corr.dropna(inplace=True)
corr = corr.join(rat['count'])
corr[corr['count']>100].sort_values('Correlation',ascending=False).head()
```
